```python
import pandas as pd
import matplotlib.pyplot as plt


# Membaca file Excel Facttable
data = pd.read_excel('FactTable.xlsx')


# Melihat ringkasan data
data.head()
```

### A.	Informasi Segmentasi Bisnis dengan Analisis Jumlah Transaksi Dalam bentuk Income Berdasarkan Tahun!

```python
import matplotlib.ticker as ticker

# Mendapatkan informasi pendapatan berdasarkan tahun
income_by_year = data.groupby('Year')['Income'].sum()

# Mengatur warna bar berdasarkan perubahan pendapatan
colors = ['green' if income_by_year.iloc[i] > income_by_year.iloc[i - 1] else 'red' for i in range(len(income_by_year))]

# Visualisasi pendapatan berdasarkan tahun dengan bar warna dinamis
plt.figure(figsize=(9, 6))
bars = plt.bar(income_by_year.index, income_by_year.values, color=colors)
plt.title(' Analisis Jumlah Transaksi Dalam bentuk Income Berdasarkan Tahun')
plt.xlabel('Tahun')
plt.ylabel('Income')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Menambahkan legenda untuk bar
increased_patch = plt.Line2D([0], [0], marker='s', color='w', label='Pendapatan Meningkat', markerfacecolor='green', markersize=10)
decreased_patch = plt.Line2D([0], [0], marker='s', color='w', label='Pendapatan Menurun', markerfacecolor='red', markersize=10)
plt.legend(handles=[increased_patch, decreased_patch])

# Menambahkan simbol dollar ($) pada label sumbu y
formatter = ticker.FormatStrFormatter('$%1.2f')
plt.gca().yaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.show()
```

### F. Analisis Cost Accounting: Tren Rata-Rata Margin Keuntungan dari Waktu ke Waktu!

```python
# Menghitung Total Biaya Produksi (Cost)
data['Total_Cost'] = data['Quantity'] * data['Price']

# Menghitung Pendapatan Bersih (Net Income)
data['Net_Income'] = data['Income'] - data['Total_Cost']

# Menghitung Margin Keuntungan (Profit Margin)
data['Profit_Margin'] = data['Net_Income'] / data['Income']

# Menampilkan hasil perhitungan
print("Total Biaya Produksi (Cost):")
print(data['Total_Cost'])

print("\nPendapatan Bersih (Net Income):")
print(data['Net_Income'])

print("\nMargin Keuntungan (Profit Margin):")
print(data['Profit_Margin'])
```

```python
if 'Year' in data.columns:
    plt.figure(figsize=(10, 8))
    profit_margin_by_year = data.groupby('Year')['Profit_Margin'].mean()
    plt.plot(profit_margin_by_year.index, profit_margin_by_year.values, marker='o', linestyle='-', color='green', label='Rata-Rata Margin Keuntungan')
    plt.title('Tren Rata-Rata Margin Keuntungan dari Waktu ke Waktu')
    plt.xlabel('Tahun')
    plt.ylabel('Rata-Rata Margin Keuntungan')
    plt.grid(True)

    # Menambahkan label pada titik data
    for i, txt in enumerate(profit_margin_by_year.values):
        plt.annotate(f'{txt:.2f}', (profit_margin_by_year.index[i], txt), textcoords="offset points", xytext=(0, 10), ha='center')

    # Menambahkan legenda
    plt.legend()

    # Menentukan rentang sumbu x dan y
    plt.xlim(min(profit_margin_by_year.index) - 0.5, max(profit_margin_by_year.index) + 0.5)
    plt.ylim(min(profit_margin_by_year.values) - 0.05, max(profit_margin_by_year.values) + 0.05)

    plt.tight_layout()
    plt.show()
```
